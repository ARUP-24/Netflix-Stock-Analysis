```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
from datetime import datetime
```

```python
data = pd.read_csv('NFLX.csv')
```

```python
data
```

```python
data.head(5)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
data.head()
```

```python
data
```

Volume of stock stock VS Time

```python
data.plot(y='Volume',ylabel='Volume')
plt.title('volume of stock vs time')
plt.show()
```

Netflix stock price - high , open , close, low

```python
data.plot(y=['High','Close','Open','Low'],ylabel='Stock')
plt.title('volume of stock vs time')
plt.show()
```

Netflix Stock price - DAY , MONTTH , YEAR

```python
fig,(x1,x2,x3) = plt.subplots(3,figsize=(15,10))
data.groupby(data.index.day).mean().plot(y='Volume',ax=x1,xlabel='Days')
data.groupby(data.index.month).mean().plot(y='Volume',ax=x2,xlabel='Months')
data.groupby(data.index.year).mean().plot(y='Volume',ax=x3,xlabel='Year')
plt.show()
```

TOP 5 Dates with highest stock price

```python
a = data.sort_values(by=['High'],ascending=True).head(5)
a[['High','Low']]
```

```python
a.plot(y='High')
plt.show()
```

Top 5 with lowest stock price

```python
b = a.sort_values(by=['Low'],ascending=True)
b['Low']
```

```python
b.plot(y='Low')
plt.show()
```

```python
fig,axes1 = plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(20,10))
data.plot(ax=axes1[0],y='High',color='green')
data.plot(ax=axes1[1],y='Low',color='red')
```
